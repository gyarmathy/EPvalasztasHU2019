## Adatok beimportálása

```python
excel_filename_in = "datain/településeredmények_EP2019.xls"
```

```python
column_names = ["Település", 
                "Település típusa", 
                "Szavazókör db", 
                "Feldolg.%", 
                "Névjegyzékben szereplő választópolgárok száma", 
                "Megjelentek száma", 
                "Bélyegző nélküli szavazólapok száma", 
                "Lebélyegzett szavazólapok száma", 
                "Érvénytelen, lebélyegzett szavazólapok száma", 
                "Érvényes szavazólapok száma", 
                "MSZP-PÁRBESZÉD", "MKKP", "JOBBIK", "FIDESZ", "MOMENTUM", "DK", "MI HAZÁNK", "MUNKÁSPÁRT", "LMP",
               ]

megyek = ['Budapest',
         'Baranya',
         'Bács-Kiskun',
         'Békés',
         'Borsod-Abaúj-Zemplén',
         'Csongrád',
         'Fejér',
         'Győr-Moson-Sopron',
         'Hajdú-Bihar',
         'Heves',
         'Jász-Nagykun-Szolnok',
         'Komárom-Esztergom',
         'Nógrád',
         'Pest',
         'Somogy',
         'Szabolcs-Szatmár-Bereg',
         'Tolna',
         'Vas',
         'Veszprém',
         'Zala']
```

```python
kozsegi_eredmenyek = pd.DataFrame()
for sheet_num in range(1, 21):
    megye_eredmenyek = pd.read_excel(excel_filename_in,
                                 sheet_name=sheet_num,
                                 header=4, 
                                 )
    megye_eredmenyek.columns = column_names
    megye_eredmenyek["Megye"] = megyek[sheet_num-1] ## megyenév hozzáadása
    megye_eredmenyek = megye_eredmenyek.iloc[:-1, :] ## az utolsó sor a megyei összesítés
    kozsegi_eredmenyek = pd.concat([kozsegi_eredmenyek, megye_eredmenyek], ignore_index=True)
```

```python
## oszlopok átrendezése
del kozsegi_eredmenyek["Feldolg.%"] ## mindenhol 100%
kozsegi_eredmenyek = kozsegi_eredmenyek.loc[:, ["Település", "Megye", "Település típusa"]+[c for c in kozsegi_eredmenyek.columns if c not in ["Település", "Megye", "Település típusa"]]]
```
